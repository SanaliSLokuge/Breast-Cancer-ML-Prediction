```python
#To identify the outliers according to interquartile ranges
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers
```

Reused from the Code Reuse Session 1

```python
#To identify the outliers in Age column
outliers =detect_outliers_iqr(data,'Age')
print("number of outliers: "+ str(len(outliers)))
outliers
```

```python
#To drop the outliers in the Age column
data.drop(outliers.index, inplace=True)
#To ensure that the outliers are correctly dropped in Age column
outliers = detect_outliers_iqr(data,'Age')
print("number of outliers: "+ str(len(outliers)))
```

```python
#To replace the null values with the mean value
Mean_Tumor_Size = data['Tumor_Size'].mean()
Mean_Regional_Node_Examined = data['Regional_Node_Examined'].mean()
Mode_Sex = data['Sex'].mode()[0]
data['Tumor_Size'] = data['Tumor_Size'].fillna(Mean_Tumor_Size)
data['Regional_Node_Examined'] = data['Regional_Node_Examined'].fillna(Mean_Regional_Node_Examined)
data['Sex'] = data['Sex'].fillna(Mode_Sex)
Mean_Age = data['Age'].mean()
data['Age']=data['Age'].fillna(Mean_Age)
```

```python
# Create a dictionary to hold encoders for each column
encoders = {}

categorical_columns = ['Sex', 'T_Stage', 'N_Stage', '6th_Stage',
                       'Differentiated', 'A_Stage', 'Estrogen_Status',
                       'Progesterone_Status', 'Mortality_Status']

# Apply encoding and store encoder for each column
for column in categorical_columns:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
    encoders[column] = le
```
